```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Employee Dataset
data = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Finance'],
    'Salary': [50000, 60000, 55000, 70000, 65000]
}
df = pd.DataFrame(data)

# ----------------------
# 1. Bar Chart - Average Salary by Department
# ----------------------
avg_salary = df.groupby("Department")["Salary"].mean()

plt.figure(figsize=(7,5))
plt.bar(avg_salary.index, avg_salary.values, color=['skyblue','orange','green'])
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()

# ----------------------
# 2. Scatter Plot - Salary vs EmployeeID
# ----------------------
plt.figure(figsize=(7,5))
for dept in df['Department'].unique():
    subset = df[df['Department'] == dept]
    plt.scatter(subset['EmployeeID'], subset['Salary'], label=dept, s=100)

plt.title("Employee Salary Distribution")
plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.legend(title="Department")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# ----------------------
# Seaborn Scatter Plot with regression line
# ----------------------
plt.figure(figsize=(7,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df, palette="deep", s=80)
plt.title("Iris Dataset - Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create dataset directly (no CSV needed)
data = {
    "Day": list(range(1, 31)),
    "Temperature": [30, 32, 31, 29, 33, 34, 32, 31, 30, 29,
                    28, 27, 29, 31, 32, 33, 34, 35, 33, 32,
                    31, 30, 29, 28, 27, 28, 29, 30, 31, 32]
}
df = pd.DataFrame(data)

# Step 2: Plot Line Chart
plt.figure(figsize=(10,6))
plt.plot(df['Day'], df['Temperature'], marker='o', color='b', linestyle='-')

# Step 3: Add labels, title, and grid
plt.title("Daily Temperature Changes (Weather Dataset)")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset (Quarter and Sales)
# If using CSV: df = pd.read_csv("sales_data.csv")
# Here we'll create it directly
data = {
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Sales": [15000, 22000, 18000, 25000]
}
df = pd.DataFrame(data)

# Step 2: Plot Line Chart
plt.figure(figsize=(8,5))
plt.plot(df['Quarter'], df['Sales'], marker='o', color='blue', linestyle='-')

# Step 3: Highlight peak sales with a red marker
max_sales_idx = df['Sales'].idxmax()
plt.scatter(df['Quarter'][max_sales_idx], df['Sales'][max_sales_idx],
            color='red', s=120, label="Peak Sales")

# Step 4: Add labels, title, grid
plt.title("Quarterly Sales Performance (Retail Sales Dataset)")
plt.xlabel("Quarter")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.legend()

plt.show()
```
